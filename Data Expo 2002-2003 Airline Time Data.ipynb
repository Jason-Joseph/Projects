# Setting up files and libraries

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FormatStrFormatter, StrMethodFormatter

#Sets the viewing capability to 30 columns
pd.set_option('display.max_columns', 30)
```

```python
#Read csv files
y2 = pd.read_csv('2002.csv', encoding='ISO-8859-1', low_memory=False)
y3 = pd.read_csv('2003.csv', encoding='ISO-8859-1', low_memory=False)
airport = pd.read_csv('airports.csv', encoding='ISO-8859-1', low_memory=False)
carrier = pd.read_csv('carriers.csv', encoding='ISO-8859-1', low_memory=False)
plane = pd.read_csv('planedata.csv', encoding='ISO-8859-1', low_memory=False)
y2.head()
```

# Merge the csv files

```python
#Merge the two csv's into 1
flights = pd.concat([y2, y3])
flights.head()
```

```python
#Checking the datatypes
flights.info()
```

```python
#Checking All the available variables
flights.columns
```

# Cleaning Data & Introducing New Variables


```python
#Filtering timestamp so that it follows 24 hours
flights = flights [flights['DepTime'] <= 2359]
flights = flights [flights['ArrTime'] <= 2359]
flights = flights [flights['CRSDepTime'] <= 2359]
flights = flights [flights['CRSArrTime'] <= 2359]
```

```python
flights.shape
```

```python
#Overview of data
flights.describe()
```

```python
#Checking number of duplicate rows
print('Number of duplicate values: ', len(flights[flights.duplicated() == True]))
```

```python
#Post-Duplicate removal
flights.drop_duplicates(subset=None, inplace=True)
flights.head()
print('Number of duplicate values after cleaned: ', len(flights[flights.duplicated() == True]))

#Looking for null values
print(flights.isnull().sum())
print('Original Shape: ', flights.shape)
```

```python
#Dropping columns that are filled with only NA Values
flights.dropna(axis =1, how = 'all', inplace = True)
flights.dropna(axis = 0, how = 'all', inplace = True)
print('Shape after dropping columns with all NA: ', flights.shape) #There are no missing values
flights.head()
```

```python
#Dropping variables unrelated to the question
flights = flights.drop(['TaxiIn' , 'TaxiOut', 'Diverted'], axis = 1)
flights.head()
```

```python
#Checking number of cancelled flights
len(flights[flights['Cancelled'] == 1])
```

No cancelled flights

```python
#Introducing new variable delay
flights['delay'] = flights['ArrDelay'] + flights['DepDelay']

#Filter & Sort values of delay > 0
flights = flights[flights['delay'] > 0 ].sort_values(by = ('delay')) 
```

# Question 1
When is the best time of day, day of the week, and time of year to fly to minimise delays?

```python
#Finding average delay by Day Of Week
day = flights.groupby(flights['DayOfWeek']).mean().reset_index()
day.drop(['Year', 'Month', 'DayofMonth'], axis = 1) 
day.head()
```

```python
day[['DayOfWeek', 'delay']].sort_values('delay').head()
```

```python
#Line plot for best day
fig, ax = plt.subplots(figsize = (8,5))
plt.plot(day['DayOfWeek'], day['delay'], color = '#66b2b2')
plt.title("Average Flight Delay by Day")
plt.xlabel('Day of Week')
plt.ylabel('Average Delay (in minutes)')
```

**Best day of the week to fly is on Saturday**

```python
#Average delay based on time of departure
time = flights.groupby(flights['CRSDepTime']).mean().reset_index()
time.drop(['Year', 'Month', 'DayofMonth', 'DayOfWeek'], axis = 1)
time[['CRSDepTime', 'delay']].sort_values('delay').head(10)
```

```python
#Line plot for time of flight
fig, ax = plt.subplots(figsize = (10,5))
plt.plot(time['CRSDepTime'], time['delay'], color = '#66b2b2')
plt.title('Average Flight Delay by Time')
plt.xlabel('Departure Time')
plt.xticks([0, 500, 1000, 1500, 2000, 2359])
plt.ylabel('Average Delay (in minutes)')
```

**Best time of day to fly is at 3.30 AM**

```python
#Finding average delay for 2002
year2 = flights[flights['Year'] == 2002]
y2_month = year2.groupby(year2['Month']).mean().reset_index()
y2_month[['Month', 'Year', 'delay']].sort_values('delay').head()
```

```python
#Finding average delay for 2003
year3 = flights[flights['Year'] == 2003]
y3_month = year3.groupby(year3['Month']).mean().reset_index()
y3_month[['Month', 'Year', 'delay']].sort_values('delay').head()
```

```python
#Line chart for best time of year
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('Average Delays')

ax[0].plot(y2_month['Month'], y2_month['delay'], color = '#66b2b2')
ax[0].title.set_text('Flights in 2002')
ax[0].set_xlabel("Month")
ax[0].set_ylabel("Average Delay (in minutes)")

ax[1].plot(y3_month['Month'], y3_month['delay'], color = '#008080')
ax[1].title.set_text('Flights in 2003')
ax[1].set_xlabel("Month")
ax[1].set_ylabel("Average Delay (in minutes)")

```

# Question 2
Do older planes suffer more delays?

```python
#Count number of flights by year
num_flights = flights.merge(plane.rename(columns={'tailnum': 'TailNum'}),
                   how='left',
                   on='TailNum', 
                   copy=False).groupby(['year', 'TailNum']).mean().reset_index().sort_values('year')

num_flights = num_flights[num_flights.year != '0000']
num_flights = num_flights[num_flights.year != 'None']
num_flights = num_flights[num_flights.year != '2007']
num_flights[['year', 'delay']].sort_values('delay').head(5)
num_flights.groupby(['year'])['year'].count().reset_index(name = 'Count').sort_values('Count', ascending = False).head(10)
```

```python
#Visualization for number of flights
fig, ax = plt.subplots(figsize = (15,5))
plt.scatter(num_flights['year'], num_flights['delay'], color = '#88d8b0')
plt.xticks(rotation=90)
plt.title('Number of Delays by Manufacturing Years')
plt.xlabel('Manufacturing Year', labelpad = 15)
plt.ylabel('Average Delay (in minutes)')
plt.show()
```
